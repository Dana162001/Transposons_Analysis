```python
import matplotlib.pyplot as plt
from matplotlib.patches import ConnectionPatch
import numpy as np
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Inne', 'Geny kodujące transpozazy', 'Geny oporności na antybiotyki'
sizes = [18132-12223-438, 12223, 438]
explode = (0.05, 0.05, 0.1)  # only "explode" one slice 
colors = ['deepskyblue','magenta','aqua']

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, colors=colors, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=360)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# make figure and assign axis objects
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 5))
fig.subplots_adjust(wspace=0)

# pie chart parameters
overall_ratios = [438, 12223, 18132-12223-438]
labels = ['Geny oporności\n na antybiotyki', 'Geny kodujące\n transpozazy', 'Inne']
explode = [0.1, 0, 0]
colors = ['lightblue','lightgreen','violet']

# rotate so that first wedge is split by the x-axis
angle = -180 * overall_ratios[0]
wedges, *_ = ax1.pie(overall_ratios, autopct='%1.1f%%', startangle=angle,
                     labels=labels, explode=explode, colors=colors)

# bar chart parameters
arg_ratios = [.466,.461,.073]
arg_labels = ['ANT(9)Ia - oporność\n na aminoglikozydy', 'Erm - oporność\n na erytromycynę', 'SAT-4 - oporność\n na streptotrycynę']
bottom = 1
width = .2

# Adding from the top matches the legend.
for j, (height, label) in enumerate(reversed([*zip(arg_ratios, arg_labels)])):
    bottom -= height
    bc = ax2.bar(0, height, width, bottom=bottom, color='C0', label=label,
                 alpha=0.1 + 0.25 * j)
    ax2.bar_label(bc, labels=[f"{height:.0%}"], label_type='center')

ax2.set_title('Rozkład genów oporności na antybiotyki')
ax2.legend()
ax2.axis('off')
ax2.set_xlim(- 2.5 * width, 2.5 * width)

# use ConnectionPatch to draw lines between the two plots
theta1, theta2 = wedges[0].theta1, wedges[0].theta2
center, r = wedges[0].center, wedges[0].r
bar_height = sum(arg_ratios)


plt.legend(bbox_to_anchor=(0.8, 0.9), loc='upper left', borderaxespad=0, fontsize=12)
plt.show()
```
