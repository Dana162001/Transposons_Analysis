```python
import pandas as pd
```

```python
path_fmt = '/home/dana/Desktop/p_sa/results/blastp/20/res_fmt_len.csv'
```

```python
df = pd.read_table(path_fmt, header=None, names=["qseqid","qstart","qend", "qlen","sseqid", "sstart", "send", "slen","evalue", "ppos"])
```

```python
df
```

```python
df['qcovs'] = ((df['qstart'] - df['qend']).abs() + 1) / df['qlen']
df['scovq'] = ((df['sstart'] - df['send']).abs() + 1) / df['slen']
df
```

```python
new_df = df.sort_values(['qcovs','scovq','ppos','evalue'], ascending=[False, False, False, True]).copy()
new_df
```

```python
new_df.drop_duplicates('qseqid', inplace=True)
new_df
```

```python
new_df = new_df[ (new_df['qcovs'] >= 0.8) & (new_df['scovq'] >= 0.8) & (new_df['evalue'] <= 0.0001) ]
#new_df = new_df[ (new_df['qcovs'] >= 0.2) & (new_df['scovq'] >= 0.5) & (new_df['evalue'] <= 0.0001) ]
new_df
```

```python
df['slen_2'] = df.apply(lambda x: x['send'] - x['sstart'], axis=1)
df['qlen_2'] = df.apply(lambda x: x['qend'] - x['qstart'], axis=1)
df
```

```python
column = df["qlen_2"]

max_value = column.max()

print(max_value)
```

```python
df['qcov'] = df.apply(lambda x: (x['qlen_2']+1)/x['qlen'], axis=1)
df['scov'] = df.apply(lambda x: (x['slen_2']+1)/x['slen'], axis=1)
df
```

```python
df_sort_qcov = df[df['qcov'] > 0.8]
df_sort_qcov
```

```python
df_sort_cov = df_sort_qcov[df_sort_qcov['scov'] > 0.8]
df_sort_cov
```

```python
df_sort_cov = df_sort_cov[df_sort_cov['ppos'] > 80]
df_sort_cov
```

```python
new_df = df_sort_cov[['qseqid','sseqid','qcov','scov','ppos','evalue']].copy()
new_df
```

```python
arg_id = [x[-1].lower() for x in length_of_df["sseqid"].str.split("|")]
```

```python
duplicates = ["car", "erm", "msr"]
```

```python
tmp = length_of_df.copy()

for d in duplicates:
    duplicated_ids = []
    for i in range(len(arg_id)):
        if d in arg_id[i]:
            duplicated_ids.append(i)
    if len(duplicated_ids) < 2:
        continue
    best_id = length_of_df.iloc[duplicated_ids]["ppos"].idxmax()
    other_ids = length_of_df.iloc[duplicated_ids].index.tolist()
    other_ids.remove(best_id)
    tmp = tmp.drop(other_ids)

len(tmp)
```
