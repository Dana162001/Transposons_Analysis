```python
df['slen_2'] = df.apply(lambda x: x['send'] - x['sstart'], axis=1)
df['qlen_2'] = df.apply(lambda x: x['qend'] - x['qstart'], axis=1)
df
```

```python
column = df["qlen_2"]

max_value = column.max()

print(max_value)
```

```python
df['qcov'] = df.apply(lambda x: (x['qlen_2']+1)/x['qlen'], axis=1)
df['scov'] = df.apply(lambda x: (x['slen_2']+1)/x['slen'], axis=1)
df
```

```python
dfsorted = df[df['evalue'] < 1]
```

```python
length_of_df = dfsorted.drop_duplicates(subset=['qseqid'])
length_of_df
```

```python
arg_id = [x[-1].lower() for x in length_of_df["sseqid"].str.split("|")]
```

```python
duplicates = ["car", "erm", "msr"]
```

```python
tmp = length_of_df.copy()

for d in duplicates:
    duplicated_ids = []
    for i in range(len(arg_id)):
        if d in arg_id[i]:
            duplicated_ids.append(i)
    if len(duplicated_ids) < 2:
        continue
    best_id = length_of_df.iloc[duplicated_ids]["ppos"].idxmax()
    other_ids = length_of_df.iloc[duplicated_ids].index.tolist()
    other_ids.remove(best_id)
    tmp = tmp.drop(other_ids)

len(tmp)
```
